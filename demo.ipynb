```python
import numpy as np

from methods import *
from plotting import *
```

```python
def f(x): return x - np.cos(x)
def f_grad(x): return 1 + np.sin(x)
```

```python
# Find root of f(x) by FPI on x-f(x).
xs_FPI = fpi(lambda x: x - f(x) 0.2, 0.0001)

prep_plot(xlim=(0, 1.5),
          ylim=(0, 1),
          padding=0.1)

plot_func_continuous(f_FPI)
plot_func_vals(f_FPI, xs_FPI)
```

```python
# Find root of f with Newton's method.
xs_newton = newton(f, f_grad, 0.2, 0.0001)

prep_plot(xlim=(0, 1.5),
          ylim=(-1, 1),
          padding=0.1)

plot_func_continuous(f)
plot_func_vals(f, xs_newton)
```

```python
# Approximate the solution to the differential equation y' = xy
# for the given initial conditions using Euler's method.

def g(x, y): return x * y
def y_true(x): return np.exp((x ** 2) / 2)


h = 0.2
iters = 10

x0, y0 = 0, 1

xys_euler = euler(g, (x0, y0), h, iters)
xys_euler_mod = euler_mod(g, (x0, y0), h, iters)

prep_plot(xlim=(0, 2),
          ylim=(1, 8))

plot_func_continuous(y_true)

plot_xys(xys_euler, point_colour="g")
plot_xys(xys_euler_mod, point_colour="b")
```
